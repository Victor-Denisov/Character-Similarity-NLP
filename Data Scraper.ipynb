# Data Scrape

This notebook will scrape show script data from two resources:
1. Seinfeld Scripts
https://www.seinfeldscripts.com/seinfeld-scripts.html

```python
# import necessary data scraping libraries, for this task I used selenium
import numpy
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
```

### Scrape Seinfeld Scripts

```python
# define the driver being used as a chrome web driver
driver = webdriver.Chrome()
driver.set_page_load_timeout(2)

# define list structures for stroing data
simple_list=[[]]
temp_list = []
driver.get("https://www.seinfeldscripts.com/seinfeld-scripts.html")

# Iterate across 4 td elements, otherwise known as columns listed
for i in range(1,4):
    # at each iteration, reset temp list to be empty
    temp_list = []
    # iterate across tables j, otherwise known as seasons
    for j in range(1,6):
        # iterate across k tr elements, otherwise known as the episode
        for k in range(2,30):
            # define the xpath of the episode
            x_path_user = '//*[@id="content"]/table[2]/tbody/tr/td[' + str(i) +']/table[' + str(j) + ']/tbody/tr[' + str(k) +']/td[2]/a'
            try:
                # click on the link
                driver.find_element_by_xpath(x_path_user).click()
                # find transcript and append it to temp list
                elem = driver.find_elements_by_css_selector('p')
                for el in elem:
                    temp_list.append(el.text)
                driver.back()
            except:
    # append the list of i transcripts to simple list
    simple_list.append(temp_list)
```

```python
print(simple_list[1][1])
```

```python
df = pd.DataFrame(simple_list)
```

```python
# save to csv
df.to_csv('seinfeld_transcripts.csv')
```
